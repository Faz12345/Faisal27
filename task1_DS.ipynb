```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\FAISAL-PC\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_87\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_87.csv.csv")
data.head()
```

```python
gender_employment = data.groupby(['Region', 'IncomeGroup']).size().unstack()
fig, ax = plt.subplots(figsize=(10, 6))
gender_employment.plot(kind='bar', stacked=True, ax=ax)
ax.set_xlabel('Gender and Employment Status')
ax.set_ylabel('Count')
ax.set_title('Distribution of Employment Status and Recruitment Source by Gender')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```
